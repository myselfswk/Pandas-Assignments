# Hierarchical Indexing part 1 & 2

```python
#hierarical indexing is also hyper-dimentional indexing
data = pd.Series(np.random.randn(9), index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'],
                                            [1, 2, 3, 1, 3, 1, 2, 2, 3]])
display(data)   #Hieraricahaaly indexed object partail indexing is possible
```

```python
#now, we can access the whole chunk by just calling upper index like
display(data['b'])
print("----------------------")
display(data['a' : 'b'])   #selecting from one index to another
```

```python
#you can also access this data by a method called loc[]
display(data.loc[['b', 'd']])   #it doesn't give you range of the element you select
#loc gives you the inner values of the selected element
print("--------------------")
#I can access inner level of index, which will select all upper index at level 1
display(data.loc[:, 2])  #[see all values from dimention 1, only selected values from dimention '2']
#all those upper indexes that contain '2' index will be printed
```

```python
#Hierarchical indexing plays an important role in reshaping data and group-based operations like forming a pivot table.
#For example, you could rearrange the data into a DataFrame using its unstack method:
#pivot simply means reorganize of data
display(data)
print("------------------")
data_unstack = data.unstack()   #those indexes which don't contain values will be NaN
display(data_unstack)
print("---------------------")
#you can also convert your unstack data into stack again
data.unstack().stack()
```

```python
#With a DataFrame, either axis can have a hierarchical index, i-e [18]
data_frame = pd.DataFrame(np.arange(12).reshape((4, 3)),
                         index=[['a', 'a', 'b', 'b'], [1, 2, 1, 2]],  #1, 2 is the inner index of data_frame index
                         columns=[['Karachi', 'Karachi', 'Lahore'], ['Green', 'Blue', 'Green']])
display(data_frame)
print("--------------------")
#we can also gives indexes and columns some names, like
data_frame.index.names = ['Key1', 'Key2']
data_frame.columns.names = ['Cities:', 'Colors:']  #color is the inner index of data_frame columns
display(data_frame)
```

```python
display(data_frame['Karachi'])  #gives us all the data relate with 'Karachi'
#Be careful to distinguish the index names 'Cities' and 'colors', simply 'Cities' & 'color' aren't relate with each other
#from the row labels, inner functionalities can be handle by pandas itself
```

```python
#@@  Reordering and sorting Levels
data_frame.swaplevel('Key1', 'Key2')  #but there is no inner and outer index right now
#can save in the same data frame variables or another variables like
# frame = data_frame.swaplevel('key1', 'key2')
```

```python
#sort_index , on the other hand, sorts the data using only the values in a single level.
#When swapping levels, it’s not uncommon to also use sort_index so that the result is
#lexicographically sorted by the indicated level:
display(data_frame.sort_index(level=1))
print("--------------------------------")
# compare the above statement output with following statement output
data_frame.swaplevel(0, 1).sort_index(level=0)
```

# Combining and Merging Datasets

```python
#pandas.merge connects rows in DataFrames based on one or more keys.
#pandas.concat concatenates or “stacks” together objects along an axis.
#The combine_first instance method enables splicing together overlapping data to fill in missing values in one object
#with values from another.

df1 = pd.DataFrame({
    'Key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
    'Data1': range(7)
})
display("DataFrame 1", df1)
print("------------------------------")
df2 = pd.DataFrame({
    'Key': ['a', 'b', 'd'],
    'Data2': range(3)
})
display("DataFrame 2", df2)
print("After Merging DF1 & DF2, Values will be")
display(pd.merge(df1, df2, on='Key'))
print("-------------------------------")
df3 = pd.DataFrame({
    'LKey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
    'Data1': range(7)
})
display("DataFrame 3", df3)
print("-------------------------------")
df4 = pd.DataFrame({
    'RKey': ['a', 'b', 'd'],
    'Data2': range(3)
})
display("DataFrame 4", df4)
print("After Merging DF3 & DF4, Values will be")
#for merging dataframe, we always need some same columns to merge
display(pd.merge(df3, df4, left_on='LKey', right_on='RKey'))
```

```python
d_f1 = pd.DataFrame({
    'LKey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
    'Data1': range(7)
})
print("First DataFrame:")
display(d_f1)
d_f2 = pd.DataFrame({
    'RKey': ['a', 'b', 'd'],
    'Data2': range(3)
})
print("Second DataFrame:")
display(d_f2)
pd.merge(d_f1, d_f2, left_on='LKey', right_on='RKey')
```

```python
left_one = pd.DataFrame({
    'Key': ['a', 'b', 'a', 'a', 'b', 'c'],
    'Value': range(6)
})
display(left_one)
right_one = pd.DataFrame({
    'Group_value': [4, 5]
}, index = ['a', 'b']
)
display(right_one)
```

```python
#now, merging left_one and right_one
display(pd.merge(left_one, right_one, left_on='Key', right_index=True))   #inner join, it's must match the data
print("----------------------------")
display(pd.merge(left_one, right_one, left_on='Key', right_index=True, how='outer'))
```

# Concatenating Along an Axis

```python
#Another kind of data combination operation is referred to interchangeably as concat‐ enation, binding, or stacking
S1 = pd.Series([0, 1], index=['a', 'b'])
S2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])  #there is no merging in this process, only stacking
S3 = pd.Series([5, 6], index=['f', 'g'])  #there is only concatenate the values by this code
display(S1, S2, S3)
print("--------------------------------")
#By default concat works along axis=0
display(pd.concat([S1, S2, S3]))
#pd.concat([s1, s2, s3], axis=1)
#In this case there is no overlap on the other axis, which as you can see is the sorted.union (the 'outer' join) of the indexes.
print("--------------------------------")
#You can instead intersect them by passing
#join='inner' :
S4 = pd.concat([S1, S3])
display(S4)
display(pd.concat([S1, S4], axis=1, sort=True))
display(pd.concat([S1, S4], axis=1, join='inner'))
print("----------------------------------")
#In this last example, the 'f' and 'g' labels disappeared because of the join='inner' option
#You can even specify the axes to be used on the other axes with join_axes :
display(pd.concat([S1, S4], axis=1, sort=True ,join_axes=[['a', 'c', 'b', 'e']]))
```

```python
#after adding key into the series to merge them
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
result = pd.concat([s1, s1, s3], keys=['one', 'two', 'three'])
display(result)
print("-----------------------------------")
#A last consideration concerns DataFrames in which the row index does not contain any relevant data:
```

# Pivoting Long to Wide Format

```python
#here we read some csv files through built-in pandas methods
data = pd.read_csv('examples/macrodata.csv')
display(data.head())   #head() gives only top 5 data outputs
#period index method convert data into specific date
print("----------------------------")
period_time = pd.PeriodIndex(year=data.year, quarter=data.quarter, name='data')
print(type(period_time))
display(period_time)
print("---------------------------")
#if we're going to create columns from the given data then, we use
columns = pd.Index(['realgdp', 'infl', 'unemp'], name='items')
display(columns)  #here our index is created
#now we reindex our 'data' dataframe
data = data.reindex(columns=columns)   #by this line of code, data is change into our given columns
display(data.head())
print("------------------------------")
```

```python
#to_timestamp convert quarter type date into proper date
display(period_time.to_timestamp('D', 'end'))
data.index = period_time.to_timestamp('D', 'end')
print("--------------------------------")
display(data.head())
```

```python
#we can also arrange our work column wise, also this is again convert back into older format
ldata = data.stack().reset_index().rename(columns={0: 'value'})
display(ldata[:10])   #this is what you called hierarical format type data
```

```python
#An inverse operation to pivot for DataFrames is pandas.melt . Rather than transforming one column into many in a new DataFrame,
#it merges multiple columns into one, producing a DataFrame that is longer than the input.
d_f5 = pd.DataFrame({
    'Key': ['Biryani', 'Pizza', 'Burger'],
    'A': [1, 2, 3],
    'B': [4, 5, 6],   #this format is pivot format
    'C': [7, 8, 9]
})
display(d_f5)
print("-----------------------------------")
#melt() method convert pivot format into hierarchical format
melted = pd.melt(d_f5, ['Key'])
display(melted)    #this format is hierarchical format
```
